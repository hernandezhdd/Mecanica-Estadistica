# Distribucion binomial y caminante aleatorio

```python
from scipy.stats import binom

n, p = 5, 0.4

r = binom.rvs(n, p, size=1000)
# r
```

```python
import numpy as np

# np.hist(r)
```

```python
import matplotlib.pyplot as plt

plt.hist(np.array(r))
```

## 1000 caminantes aleatorios dan 1000 pasos

```python
import pandas as pd

p = 0.5
q = 1-p
n = 1000

results_1000 = np.asarray(np.random.rand(n,n) > p, dtype=int)

results_1000 = 2*results_1000 - 1

df_results_1000 = pd.DataFrame(results_1000)
```

```python
df_acum_1000 = df_results_1000.cumsum()

df_acum_1000.tail(1)
```

```python
plt.figure(figsize=(12,7))

plt.plot(df_acum_1000)

plt.plot(df_acum_1000.mean(axis=1), linewidth=4, label='Promedio')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12,7))

df_acum_1000.iloc[-1].hist(alpha=0.3, label='Paso 1000')
df_acum_1000.iloc[300].hist(alpha=0.5, label='Paso 300')
df_acum_1000.iloc[100].hist(alpha=0.8, label='Paso 100')
df_acum_1000.iloc[0].hist(alpha=1, label='Paso 1')

plt.legend()

plt.show()
```

## 100 caminantes aleatorios dan 100 pasos

```python
import pandas as pd

p = 0.3
q = 1-p
n = 100

results = np.asarray(np.random.rand(n,n) > p, dtype=int)

results = 2*results - 1

df_results = pd.DataFrame(results)
```

```python
df_acum = df_results.cumsum()

df_acum.tail(1)
```

```python
plt.figure(figsize=(12,7))

plt.plot(df_acum)

plt.plot(df_acum.mean(axis=1), linewidth=4, label='Promedio')

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12,7))

df_acum.iloc[0].hist(alpha=0.7, label='Paso 1')
df_acum.iloc[10].hist(alpha=0.7, label='Paso 10')
df_acum.iloc[30].hist(alpha=0.7, label='Paso 30')
df_acum.iloc[-1].hist(alpha=0.7, label='Paso 100')

plt.legend()

plt.show()
```
